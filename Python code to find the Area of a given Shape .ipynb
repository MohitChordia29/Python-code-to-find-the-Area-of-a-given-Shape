## Python Program to calculate the Area of Shape

#### Problem Statement:

To Write a Python program which calculates the area of various geometrical figures as given below:
- The program should ask the user about the geometrical figure for which he/she wants to calculate the area.
- Take the inputs accordingly and calculate the area. (units/metrics like mm, cm can be included as per the programmerâ€™s choice)

Desired Output:

- Which geometrical figure would you like to calculate the area for? 
- Please choose the option number:
   1. Square 
   2. Rectangle 
   3. Circle 
   4. Triangle
   5. Cylinder 
   6. Ellipse 
   7. Cone 
   8. Sphere 
   9. Parallelogram 
   10. Cube
   11. Trapezoid
   12. Torus 
 



```python
# Python code:
# Asking the user for the inputs for the desired shape to calculate the area for.
shape = int(input('Which geometrical figure would you like to calculate the area for? \nPlease choose the option number:\n 1. Square \n 2. Rectangle \n 3. Circle \n 4. Triangle \n 5. Cylinder \n 6. Ellipse \n 7. Cone \n 8. Sphere \n 9. Parallelogram \n 10. Cube \n 11.Trapezoid \n 12. Torus \n \n Enter the option:'))
print()
while True:
    # Area of Square:
    if 1==shape:
        a=int(input('Please enter the length of side'))
        aa=a*a
        print()
        print('The area of Square is:',aa)
        break
    # Area of Rectangle:
    if 2==shape:
        a=int(input('Please enter the length of side'))
        b=int(input('Please enter the breath of a side:'))
        aa=a*b
        print()
        print('The area of Rectangle is:',aa)
        break
    # Area of Circle:
    elif 3==shape:
        a=int(input('Please enter the radius of a circle:'))
        pi= 3.147
        aa= 2* pi* (a*a)
        print()
        print('The area of Circle is:',aa)
        break
    # Area of Square:
    elif 4==shape:
        a=int(input('Please enter the height of a side:'))
        b=int(input('Please enter the breath of a side:'))
        aa=0.5*(a*b)
        print()
        print('The area of Triangle is:',aa)
        break
    # Area of Cylinder:
    elif 5==shape:
        a=int(input('Please enter the Radius of a cylinder:'))
        b=int(input('Please enter the Height of a cylinder:'))
        pi= 3.147
        aa= (2* pi* (a*b)) + (2 * pi *(a*a))
        print()
        print('The area of Cylinder is:',aa, 'cm^2')
        break
    # Area of Ellipse:
    elif 6==shape:
        a=int(input('Please enter the first Radii:'))
        b=int(input('Please enter the second Radii:'))
        pi= 3.147
        aa= pi* (a*b)
        print()
        print('The area of Ellipse is:',aa, 'cm^2')
        break
    # Area of Cone:
    elif 7==shape:
        a=int(input('Please enter the Radius:'))
        b=int(input('Please enter the Side:'))
        pi= 3.147
        aa= pi* (a*b)
        print()
        print('The area of Cone is:',aa, 'cm^2')
        break
    # Area of Sphere:
    elif 8==shape:
        a=int(input('Please enter the Radius:'))
        pi= 3.147
        aa= 4*pi* (a**2)
        print()
        print('The area of Sphere is:',aa, 'cm^2')
        break
    # Area of Parallelogram:
    elif 9==shape:
        a=int(input('Please enter the base of a side:'))
        b=int(input('Please enter the height of a side:'))
        aa=a*b
        print()
        print('The area of Parallelogram is:',aa)
        break
    # Area of Cube:
    elif 10==shape:
        a=int(input('Please enter the Side:'))
        aa=6*(a**2)
        print()
        print('The area of Cube is:',aa)
        break
    # Area of Trapezoid:
    elif 11==shape:
        a1=int(input('Please enter the first base of a side:'))
        a2=int(input('Please enter the second base of a side:'))
        b=int(input('Please enter the height of a side:'))
        aa=b*((a1+a2)/2)
        print()
        print('The area of Trapezoid is:',aa)
        break
    # Area of Torus:
    elif 12==shape:
        a=int(input('Please enter the first Radii:'))
        b=int(input('Please enter the second Radii:'))
        pi= 3.147
        aa= (pi**2) *((b**2)-(a**2))
        print()
        print('The area of Torus is:',aa, 'cm^2')
        break
    # To get the correct input from the user:
    else:
        shape=int(input('Invalid input, Enter a correct option please '))
        continue
```
